```python
## removing unnecessary columns

df1 = df.copy()

## i am copying the same dataset for clarity

df1.columns
```

```python
df1.drop(['Side','Zipcode','Description','Number','Street','Timezone',
        'Airport_Code','Weather_Timestamp','Temperature(F)','Wind_Chill(F)',
       'Humidity(%)', 'Pressure(in)', 'Visibility(mi)', 'Wind_Direction',
       'Wind_Speed(mph)', 'Precipitation(in)', 'Weather_Condition', 'Amenity',
       'Bump', 'Crossing', 'Give_Way', 'Junction', 'No_Exit', 'Railway',
       'Roundabout', 'Station', 'Stop', 'Traffic_Calming', 'Traffic_Signal',
       'Turning_Loop', 'Sunrise_Sunset', 'Civil_Twilight', 'Nautical_Twilight'],axis=1,inplace=True)
```

```python
df1.shape
```

```python
## tap 10 staes with most accident cases

top_10_cities = df1['City'].value_counts().sort_values(ascending=False)[:10]
top_10_cities
```

```python
plt.figure(figsize=(6,4))
plt.xticks(rotation=90)
plt.bar(top_10_cities.index,top_10_cities.values,color='RED')
plt.title("Top 10 cities with the most Accident cases")
```

```python
## distribution of the start time

## Mostly the higher accidents occur between 3pm to 6pm approx

df1['Start_Time']=pd.to_datetime(df1['Start_Time'])
df1['Start_Time'].dt.time.value_counts().sort_index().plot()
plt.title('Distribution of Start Time')
plt.show()
```

```python
## we can also figure out the most of the accidents happening on week days or weekends

def day_type(day):
    days=list(range(0,5))
    if (day in days):
        return 'Week day'
    else:
        return 'Week end'
    
df1['Start_Time']=pd.to_datetime(df1['Start_Time'])
df1['End_Time']=pd.to_datetime(df1['End_Time'])
df1['year']=df1['Start_Time'].dt.year
df1['hour']=df1['Start_Time'].dt.hour
df1['wday_wend']=df1['Start_Time'].dt.weekday.map(day_type)
        
```

```python
pd.crosstab(df1['hour'],df1['wday_wend']).plot(rot=0,label='day')
plt.title('Distribution of accidents by hour on weedays VS weekends')
plt.ylabel('Accidents')
plt.xlabel('Hours')
plt.show()
```

```python
## what time on sunday starts on?

plt.figure(figsize=(5,3))
st_sunday=df1[df1['Start_Time'].dt.weekday==6]
st_sunday['Start_Time'].dt.hour.value_counts().sort_index().plot()
plt.title('Distribution of accidents by hour on Sundays')
```
